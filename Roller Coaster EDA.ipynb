```python
#Univariate Analysis
df['Year_Introduced'].value_counts()
#presents the frequency/count of each unique value in the set
```

```python
graph = df['Year_Introduced'].value_counts() \
.head(10).plot(kind = 'bar', title = 'Most Coasters Introduced By Year (First 10)')
graph.set_xlabel('Year_Introduced')
graph.set_ylabel('Count')
```

```python
hist = df['Speed_mph'].plot(kind = 'hist', bins = 20, title = 'Coaster Speed (mph)', color = 'red')
hist.set_xlabel('Speed (mph)')
#Get an idea of the distribution of your continuous variables
```

```python
kernal_density = df['Speed_mph'].plot(kind = 'kde', title = 'Coaster Speed (mph)', color = 'green')
kernal_density.set_xlabel('Speed (mph)')
#slightly easier to read and interpret due to normalized distributions
```

```python
#Let's observe the correlation between speed and height
df.plot(kind = 'scatter', x = 'Height_ft', y = 'Speed_mph', title = 'Coaster Height Vs Coaster Speed')
plt.show()
```

```python
sns.scatterplot(x = 'Height_ft', y = 'Speed_mph', hue = 'Year_Introduced', data = df)
```

```python
sns.pairplot(df, vars = ['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'], hue = 'Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']] \
.dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot = True)
#note, every value is in perfect correlation with itself, hence the 1s along 
#the diagonal.
```

```python
#Which locations have the fastest roller coasters (minimum of 10 coasters per location)?
```

```python
df['Location'].value_counts()
```

```python
bar_graph = df.query('Location != "Other"') \
         .groupby('Location')['Speed_mph'] \
         .agg(['mean', 'count']) \
        .query('count >= 10') \
        .sort_values('mean')['mean'] \
    .plot(kind = 'barh', title = 'Average Coaster Speed By Location', color = 'purple')
bar_graph.set_xlabel('Average Coaster Speed')
plt.show()
#We are only plotting the average speed
```
